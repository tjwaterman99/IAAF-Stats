#### Calculates the winning margin for men and women in the 100m dash

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# See the datasets folder to get this csv
df = pd.read_csv('IAAF_Results.csv')

top_two = df[(df['place'] == 1) | (df['place'] == 2)]
groups = top_two.groupby(['location','gender','distance'])

# Used to extract the winning margin
def get_margin(g):    
    location,gender,distance = g[0]
    
    first_place_time = g[1]['mark_time'].min()
    second_place_time = g[1]['mark_time'].max()
    margin = (second_place_time - first_place_time)/second_place_time
    
    results = {
        'location' : location,
        'gender' : gender,
        'distance' : distance,
        'margin' : margin
    }
    return results
    
    
margin_df = pd.DataFrame([get_margin(g) for g in groups])

mens_100_margin = margin_df[(margin_df['distance'] == 100) & (margin_df['gender'] == 'men')]
womens_100_margin = margin_df[(margin_df['distance'] == 100) & (margin_df['gender'] == 'women')]

plt.plot(mens_100_margin['margin'])
plt.plot(womens_100_margin['margin'])


plt.show()
```
